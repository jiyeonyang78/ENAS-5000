```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# need to define 4 values
# zeta, omega_0, y_0, y_dot_0 (initial conditions)
# A, B, omega_d are derived from these 4 values

zeta = 0.2 # need to be in a certain range between 0 and 1
omega_0 = 2 * np.pi
omega_d = omega_0 * np.sqrt(1 - zeta**2)
t = np.linspace(0, 3, 100)

A = 1; B = 1;
y = np.exp(-zeta * omega_0 * t) * [A*np.cos(omega_d*t) + B*np.sin(omega_d*t)]

plt.plot(t, y[0])

```
